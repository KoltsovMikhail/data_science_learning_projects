## –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–°–∫–∞—á–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç—ã train –∏ test –≤ –ª–æ–∫–∞–ª—å–Ω—É—é –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é `PATH`, –¥–∞–ª–µ–µ –∑–∞–≥—Ä—É–∑–∏–º —É–∫–∞–∑–∞–Ω–Ω—ã–µ –¥–∞—Ç–∞—Å–µ—Ç—ã:

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —á—Ç–µ–Ω–∏—è csv-—Ñ–∞–π–ª–∞ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, –≤—ã–≤–æ–¥ –æ—Å–Ω–æ–≤–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞–∏–∏ –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –Ω–∞ —ç–∫—Ä–∞–Ω.
# –í –∫–∞—á–µ—Å—Ç–≤–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
# –†–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å - ','. –î–µ—Å—è—Ç–∏—á–Ω—ã–π —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å - '.'
def read_and_display(path):
    
    df = pd.read_csv(path)
    display(df.head(5))
    df.info()
    
    return df


# –ß—Ç–µ–Ω–∏–µ —Åsv-—Ñ–∞–π–ª–æ–≤
df_train = read_and_display(PATH+'/kaggle_startups_train_01.csv')
df_test = read_and_display(PATH+'/kaggle_startups_test_01.csv')
```

## –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞

### –ü–µ—Ä–≤–∏—á–Ω–æ–µ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–æ —Å –¥–∞–Ω–Ω—ã–º–∏

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–∞ —Å –¥–∞–Ω–Ω—ã–º–∏:

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–∞ —Å –¥–∞–Ω–Ω—ã–º–∏
# name - –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
# df - –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
def data_first_look(name, df):
    print('\n–î–∞—Ç–∞—Ñ—Ä–µ–π–º:',name)
    print('–¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö:\n', df.dtypes)
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞:', df.duplicated().sum())
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:',df.isna().sum().sum())
```

```python
# —Ä–∞–∑–º–µ—Å—Ç–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –≤ —Å–ª–æ–≤–∞—Ä—å
dataframes = {
    'train': df_train,
    'test': df_test
}
# –≤—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–∞ —Å –¥–∞–Ω–Ω—ã–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
for name, df in dataframes.items():
    data_first_look(name, df)
```

–ü–µ—Ä–≤–∏—á–Ω–æ–µ –∑–Ω–∞–∫–æ–º—Å—Ç–≤–æ —Å –¥–∞–Ω–Ω—ã–º–∏ –≤—ã—è–≤–∏–ª–æ 2 —è–≤–Ω—ã—Ö –ø—Ä–æ–±–ª–µ–º—ã:

#### `—Åategory_list`

–í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ —Ä–µ—á—å –∏–¥–µ—Ç –æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ü–æ—ç—Ç–æ–º—É –Ω–∞–∏–±–æ–ª–µ–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–π –º–µ—Ç–æ–¥ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ - —ç—Ç–æ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –º–æ–¥–æ–π, —Ç–æ –µ—Å—Ç—å –º–µ—Ç–æ–¥ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏–º—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ–º.

```python
# –°–æ–∑–¥–∞–¥–∏–º –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∞ SimpleImputer —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º strategy = 'median'
imp_median = SimpleImputer(strategy = 'median')
df_train['funding_total_usd'] = imp_median.fit_transform(df_train[['funding_total_usd']])
df_test['funding_total_usd'] = imp_median.fit_transform(df_test[['funding_total_usd']])
# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ funding_total_usd –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ train', df_train.funding_total_usd.isna().sum())
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ funding_total_usd –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ test', df_test.funding_total_usd.isna().sum())
```

- –ö–∞—Ç–µ–≥–æ—Ä–∏–∏. –¢–∞–∫–∂–µ –∏–º–µ—é—Ç —Å–∏–ª—å–Ω—É—é –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –≤–Ω—É—Ç—Ä–∏ –≥—Ä—É–ø–ø—ã. –î–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ `category_1`.

–ü–æ–¥—Ä–æ–±–Ω–æ –ø—Ä–æ –∫–∞–∂–¥—ã–π –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –æ—Ç–±–æ—Ä–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –Ω–∞–ø–∏—Å–∞–Ω–æ –≤ —Å—Ç–∞—Ç—å–µ https://habr.com/ru/articles/550978/

## –í—ã–±–æ—Ä –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–†–∞–∑–¥–µ–ª–∏–º –≤—ã–±–æ—Ä–∫—É –Ω–∞ —Ç–µ—Å—Ç–æ–≤—É—é –∏ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é.

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_train.drop(['status'], axis=1),
    df_train['status'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = df_train['status'])
```

<div class="alert alert-block alert-success">

–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –≤—ã–ø–æ–ª–Ω–µ–Ω–æ –∏ –º–æ–ª–æ–¥–µ—Ü, —á—Ç–æ –∑–∞—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–ª random_state. –≠—Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤.
</div>

–ó–∞–∫–æ–¥–∏—Ä—É–µ–º —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Å –ø–æ–º–æ—â—å—é 0 –∏ 1.

```python
y_train = y_train.map({'operating': 0, 'closed': 1}).astype(int)
y_test = y_test.map({'operating': 0, 'closed': 1}).astype(int)
```

–†–∞–∑–¥–µ–ª–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ. 

```python
ohe_columns = ['country_code']
num_columns = ['funding_total_usd', 'funding_rounds', 'lifetime']
```

<div class="alert alert-block alert-success">

–û–∫, –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ —á–∏—Å–ª–æ–≤—ã–µ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã.
</div>

```python
# —Å–æ–∑–¥–∞—ë–º –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ohe_columns: OHE-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ
# –æ–≥—Ä–∞–Ω–∏—á–∏–º –∫–æ–ª–∏—á–µ—Ç—Å–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–π 10
ohe_pipe = Pipeline(
   [
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False, max_categories=10)
        )
    ]
)
```

```python
# –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('num', StandardScaler(), num_columns)    
    ], 
    remainder = 'passthrough'
)
print(data_preprocessor)
```

```python
pipe_final= Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)

param_grid = [
     #—Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 8),
        'models__max_features': range(2, 8),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
   {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 8),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

   # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE)],       
        'models__C': range(1, 8),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
   # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
       'models': [SVC(random_state=RANDOM_STATE)],
      'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  

    },
]

grid = GridSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='f1',
    n_jobs=-1,
    error_score = 'raise'
)
grid.fit(X_train, y_train)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', grid.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', grid.best_score_.round(2))
```

<div class="alert alert-block alert-success">

–ú–æ–ª–æ–¥–µ—Ü, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å pipeline. –≠—Ç–æ –¥–æ–≤–æ–ª—å–Ω–æ —É–¥–æ–±–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç. üëçüëçüëç
    
–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å: https://www.superannotate.com/blog/how-to-optimize-machine-learning-pipeline
</div>

<div class="alert alert-block alert-success">

–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–∞—Ç–µ—Ä–∏–∞–ª—ã –ø–æ –ø–æ–¥–±–æ—Ä—É –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: https://habr.com/ru/articles/563494/

–ü—Ä–æ–≤–µ—Ä–∏–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.

```python
threshold = 0.5
predict_proba = grid.predict_proba(X_test)
predicted = (predict_proba[:,1] >= threshold).astype('int')
print('–ú–µ—Ç—Ä–∏–∫–∞ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:',f1_score(y_test, predicted).round(2))
```

**–í—ã–≤–æ–¥—ã:** –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —Ä–∞–±–æ—Ç—ã –ø–∞–π–ø–ª–∞–π–Ω–∞ –±—ã–ª–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –ª—É—á—à–∞—è –º–æ–¥–µ–ª—å - DecisionTreeClassifier(max_depth=7,  max_features=2, random_state=42). –ï—ë –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å f1 –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ - 0.67, –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π - 0.66.

## –ê–Ω–∞–ª–∏–∑ –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏

–ü–æ–ª—É—á–∏–º –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è ColumnTransformer.

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è –ø–æ–ª–µ–π –∏–∑ ColumnTrasnformer
def get_column_names_from_ColumnTransformer(column_transformer):    
    col_name = []
    for transformer_in_columns in column_transformer.transformers_[:-1]:
        raw_col_name = transformer_in_columns[2]
        if isinstance(transformer_in_columns[1],Pipeline): 
            transformer = transformer_in_columns[1].steps[-1][1]
        else:
            transformer = transformer_in_columns[1]
        try:
            names = transformer.get_feature_names_out()
        except AttributeError: 
            names = raw_col_name
        if isinstance(names,np.ndarray): 
            col_name += names.tolist()
        elif isinstance(names,list):
            col_name += names    
        elif isinstance(names,str):
            col_name.append(names)
    return col_name
```

```python
# –ü–æ–ª—É—á–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
new_cat_cols = get_column_names_from_ColumnTransformer(grid.best_estimator_.named_steps['preprocessor'])
# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫ —Å–ø–∏—Å–∫—É
all_cols = np.concatenate([new_cat_cols, num_columns])
```

–ü–æ–ª—É—á–∏–º —Ç–∞–±–ª–∏—Ü—É –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –ø–æ —É–±—ã–≤–∞–Ω–∏—é.

```python
feature_importances = pd.DataFrame(
    {
        'Feature': all_cols,
        'Importance': np.abs(grid.best_estimator_._final_estimator.feature_importances_)
    })

# —Å–æ—Ä—Ç–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É –ø–æ –≤–∞–∂–Ω–æ—Å—Ç–∏ ‚Äî –æ—Ç –±–æ–ª—å—à–µ–π –∫ –º–µ–Ω—å—à–µ–π
print(feature_importances.sort_values('Importance', ascending=False))
```
